```python
from pathlib import Path
import pandas as pd
```

## Loading the data


```python
data_path = Path(r"QuestionDataV2_06122025_classified.csv").resolve()
df = pd.read_csv(data_path)
print(df.columns)
```

## Inspecting


```python
df.head()
```

```python
columns_to_keep = [
    "Question Title",
    "question.html",
    "server.js",
    "solution.html",
    "server.py",
    "properties.js",
    "info.json",
    "isAdaptive",
]
df = df[columns_to_keep]
df.head()
```

### IsAdaptive Check


```python
row_num, col_num = df.shape
adaptive_questions_count = df["isAdaptive"].sum()
print(
    f"""The number of adaptive questions is {adaptive_questions_count} \n while non adaptive questions are {row_num-adaptive_questions_count}
"""
)
py_filtered = df[(df["isAdaptive"]) & (df["server.py"].notna())]
js_filtered = df[(df["isAdaptive"]) & (df["server.js"].notna())]
```

```python
import re
import json
pattern = r"</pl-[a-z\-]+>"
question_examples = {}
for text in df["question.html"]:
    result = re.findall(pattern, text)
    for r in result:
        if not question_examples.get(r, None):
            question_examples[r] = text

save_path = Path(r"example_questions.json").resolve()
print(save_path)
save_path.write_text(json.dumps(question_examples))
print(question_examples)
```

## Notes

- info.json: Originally held question metadata, now empty. Not essential for generation but should be repopulated.
